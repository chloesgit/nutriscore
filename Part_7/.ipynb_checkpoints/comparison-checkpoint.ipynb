```python
import pandas as pd
```

```python
df = pd.read_csv("all_results.csv")
```

```python
df[['Original_Nutriscore','part_4Neo','part_4Nei','part_5','part_6']]
```

```python
df['Original_Nutriscore'].replace(['a','b','c','d','e'],
                        [1,2,3,4,5], inplace=True)
df['part_4Neo'].replace(['a','b','c','d','e'],
                        [1,2,3,4,5], inplace=True)
df['part_4Nei'].replace(['a','b','c','d','e'],
                        [1,2,3,4,5], inplace=True)
df['part_5'].replace(['a','b','c','d','e'],
                        [1,2,3,4,5], inplace=True)
df['part_6'].replace(['a','b','c','d','e'],
                        [1,2,3,4,5], inplace=True)

```

```python
df[['Original_Nutriscore','part_4Neo','part_4Nei','part_5','part_6']]
```

```python
# Print the Confusion Matrix and slice it into four pieces

import seaborn as sns
import matplotlib.pyplot as plt    
from sklearn.metrics import confusion_matrix
def comparison(org, pred,title):
    cm = confusion_matrix(org, pred)
    ax= plt.subplot()
    sns.heatmap(cm, annot=True, fmt='g', ax=ax);  #annot=True to annotate cells, ftm='g' to disable scientific notation

    # labels, title and ticks
    ax.set_xlabel('Predicted Nutriscore');ax.set_ylabel('Original Nutriscore'); 
    ax.set_title(title); 
    ax.xaxis.set_ticklabels(['a','b','c','d','e']); ax.yaxis.set_ticklabels(['a','b','c','d','e']);


```

## Part4

```python
comparison(df['Original_Nutriscore'],df['part_4Neo'],'Neo model')
```

```python
comparison(df['Original_Nutriscore'],df['part_4Nei'],'Nei model')
```

## Part5

```python
comparison(df['Original_Nutriscore'],df['part_5'],'Electre Tre model')
```

## Part 6

```python
comparison(df['Original_Nutriscore'],df['part_6'],'ML model')
```
