```python
import category_encoders as ce
# encode variables with ordinal encoding

#encoder = ce.OrdinalEncoder(cols=['product_name_fr','brands','off:nutriscore_grade'])
encoder = ce.OrdinalEncoder(cols=['Nutriscore'])
y = encoder.fit_transform(y)
# X_train = encoder.fit_transform(X_train)

# X_test = encoder.transform(X_test)
```

```python
y.value_counts()
```

```python
y
```

```python
y_org = encoder.inverse_transform(y)
```

### split into train, test data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42, stratify = y)
```

```python
X_test.shape
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

### Decision Tree

```python
# import DecisionTreeClassifier

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
from sklearn.tree import DecisionTreeClassifier

best_acc = 0

for criterion in "gini", "entropy":
    for max_depth in [2,3,4,5,6]:
        for min_samples_leaf in [5, 10, 20, 30]:
            dtree = DecisionTreeClassifier(max_depth=max_depth, criterion=criterion, min_samples_leaf=min_samples_leaf,random_state=42)
            dtree.fit(X_train, y_train)
            predicitions = dtree.predict(X_test)
            acc = accuracy_score(y_test, predicitions)
            if acc > best_acc:
                best_params = f"criterion: {criterion}, max_depth: {max_depth}, min_samples_leaf: {min_samples_leaf}"
                best_acc = acc
```

```python
best_acc
```

```python
best_params
```

```python
clf_gini = DecisionTreeClassifier(criterion='entropy', max_depth=6, random_state=42,min_samples_leaf=5)


# fit the model
clf_gini.fit(X_train, y_train)
```

```python
X.columns.values.tolist()
```

```python
from sklearn.tree import export_text 
r = export_text(clf_gini, feature_names=X.columns.values.tolist()) 
print(r)
```

```python
y_pred_gini = clf_gini.predict(X_test)
y_pred_gini
```

```python


print('Model accuracy score with criterion gini index: {0:0.4f}'. format(accuracy_score(y_test, y_pred_gini)))
```

```python
y_pred_train_gini = clf_gini.predict(X_train)

y_pred_train_gini
```

```python
print('Training-set accuracy score: {0:0.4f}'. format(accuracy_score(y_train, y_pred_train_gini)))
```

```python
# print the scores on training and test set, check overfitting 

print('Training set score: {:.4f}'.format(clf_gini.score(X_train, y_train)))

print('Test set score: {:.4f}'.format(clf_gini.score(X_test, y_test)))
```

### Confusion matrix

```python
# Print the Confusion Matrix and slice it into four pieces

import seaborn as sns
import matplotlib.pyplot as plt    
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred_gini)

print('Confusion matrix\n\n', cm)

ax= plt.subplot()
sns.heatmap(cm, annot=True, fmt='g', ax=ax);  #annot=True to annotate cells, ftm='g' to disable scientific notation

# labels, title and ticks
ax.set_xlabel('Predicted labels');ax.set_ylabel('True labels'); 
ax.set_title('Confusion Matrix'); 
ax.xaxis.set_ticklabels(['a','b','c','d','e']); ax.yaxis.set_ticklabels(['a','b','c','d','e']);


```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred_gini))
```
